### Define the functions
Define the functions that will be used across the notebook

```python
# Function to use through out the program

#identify outliers  using IQR method
def outliers_iqr(ys):
    quartile_1, quartile_3 = np.percentile(ys, [25, 99])# originally it was 25,75, adjusted to avoid removing many records
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5) 
    upper_bound = quartile_3 + (iqr * 1.5) 
    return np.where((ys > upper_bound) | (ys < lower_bound))
    #return np.where((ys > upper_bound) )
    
    
def ConvertSpeed(df, ColName):
#     filter_values = [0,25, 50,150]
#     filter_values = [0,10, 20,30,40,50,60]
    newcol = ColName+'_bin'
#     df[newcol] = pd.cut(df[ColName], bins = filter_values, labels= ["Slow","Midium","Fast"])
    df[newcol] = pd.cut(df[ColName], bins = 10, labels= range(1,11))
    return df


def timeOfDay(triptimestamp):
    hour = triptimestamp.hour
    if hour >= 5 and hour <12: return 'Morning'
    elif hour >= 12 and hour <17: return 'Afternoon'
    elif hour >= 17 and hour <21: return 'Evening'
    else :return 'Night'
    
    
def OneHotEncoderfunc(df, colList):
    c2 = []
    for c in colList:
        if df[c].nunique()>2 :
            c2.append(c)
            #c3[c] = 'ohe_'+c
    print(f'Before : {df.shape}')
    df = pd.get_dummies(df,columns=c2,drop_first=False)#prefix=c3)
    print(f'After : {df.shape}')
    return df
```

### Identify More Data Issues
 - We will assume that the wait_time_in_journey is part of the trip duration_time. therefore the wait_time_in_journey must always be less than or equals duration_time
 - We will assume that any not-cancelled trips should have a valid duration_time (not 0)
 - Another assumption is when the trip is_cancelled is true , the duration_time should be 0 
 - All time and distance should be 0 minutes/KM or more
 - Let's validate those assumption and correct the issues if any

```python
# add Training and Test to a list foe easier handling
data_combined = [trip_train_clean, trip_test]
```

```python
# make sure all distance and time columns are => 0
time_dist_cols = ['estimated_distance', 'distance_travelled', 'distance_travelled_while_moving', 'estimated_duration', 'duration_time','wait_time_initial', 'wait_time_in_journey']
for dataset in data_combined:
    for col in time_dist_cols:
        dataset.loc[dataset[col] <0, col]=0
```

```python
# Validate the assumption that wait_time_in_journey must be less than duration_time
trip_train_clean[((trip_train_clean['wait_time_in_journey'] > trip_train_clean['duration_time'])| (trip_train_clean['duration_time']==0)) & (trip_train_clean['is_cancelled'] ==0)][['duration_time','wait_time_initial','wait_time_in_journey','price','is_cancelled']]
```

 - Apparently, some data issues exist and need to be fixed

```python
# Create new temp column for duration_time 
for dataset in data_combined:
    dataset['duration_time_new']=dataset['duration_time']
```

```python
# prepare a mapping dataset for duration_time based on price and wait_time_in_journey columns
correct_trips = trip_train_clean[((trip_train_clean['wait_time_in_journey'] < trip_train_clean['duration_time'])& (trip_train_clean['duration_time']!=0)) & (trip_train_clean['is_cancelled'] ==0)][['duration_time','wait_time_in_journey','price']]
correct_trips['price']=correct_trips['price'].astype(int)
temp_avg_time = correct_trips.groupby(['price','wait_time_in_journey']).mean().rename(columns={'duration_time':'avg_duration_time'}).reset_index()
temp_avg_time
```

```python
# prepare a function to be called with each row withincorrect duration_time in training and test to fix the dara issue if exist
def get_duration_time_median(price, wait_time_in_journey):
   return int(temp_avg_time[(temp_avg_time['price'] >= price-2) & (temp_avg_time['price'] < price+2) \
                             & (temp_avg_time['wait_time_in_journey'] >= wait_time_in_journey-2)\
                            & (temp_avg_time['wait_time_in_journey'] <  wait_time_in_journey+2)]['avg_duration_time'].mean())
```

```python
# Apply the fix, fill in the incorrect duration_time with mean of duration time of other trips with similar price +-2 and similar  wait_time_in_journey +-2
for dataset in data_combined:
    dataset.loc[((dataset['wait_time_in_journey'] > dataset['duration_time'])| (dataset['duration_time']==0)) & (dataset['is_cancelled'] ==0),'duration_time_new']=\
    dataset[((dataset['wait_time_in_journey'] > dataset['duration_time'])| (dataset['duration_time']==0)) & (dataset['is_cancelled'] ==0) ].apply(lambda row: get_duration_time_median(row['price'],row['wait_time_in_journey']), axis=1)
```

```python
for dataset in data_combined:
    dataset['duration_time_new']=dataset['duration_time_new'].astype(int)
```

```python
# check the results after applying the fix
trip_train_clean[(trip_train_clean['wait_time_in_journey'] > trip_train_clean['duration_time']) & (trip_train_clean['is_cancelled'] ==0) ][['duration_time','duration_time_new','wait_time_initial','wait_time_in_journey','price','is_cancelled']]
```

 - Now duration_time_new looks better and always the wait_time_in_journey is less than duration_time

```python
# drop the exisitg column duration_time then rename the new column to duration_time
for dataset in data_combined:
    dataset.drop(columns = ['duration_time'], inplace=True, index=1)
    dataset.rename(columns={'duration_time_new':'duration_time'}, inplace=True)
```

## 3- Features Generation
 - Extract features for Training & Test datasets from existing columns
 - Adjust existing feature for better utilization

 - create a feature for missing values
 - avpid filling NaNs before feature generations
 - Rated last trip
 - rated_trips_count
 - lst_mnth_trip_cnt
 - avg_rating , - average rating , - rating count - indicator - for driver - for customer
 - request a trip price =1 if estimated_price is not empty

```python
def rated_only_count(x):
    rated_count = x[x != 0].count()
    return  rated_count

def rated_only_mean(x):
    rated_mean=x[x != 0].mean()
    return  rated_mean
summary_set = []    
for col in ['driver_id','customer_id']:    
    rate_smry = trip_train_clean[[col,'was_rated','rating']].groupby([col]).agg({'was_rated':'count','rating':[rated_only_count,rated_only_mean,'mean']}).reset_index()
    rate_smry.columns= [col, col+'_'+'trip_count', col+'_'+'rated_trip_count',col+'_'+'rated_trip_avg',col+'_'+'all_trip_rate_avg']
#     rate_smry    

    #summary_old= trip_rating[[col,'rating']].groupby([col]).agg({'rating':['count','mean']}).reset_index()
    summary_old = trip_rating[[col,'rating']].groupby([col]).agg({'rating':['count',rated_only_count,rated_only_mean,'mean']}).reset_index() 
    summary_old.columns =[col, col+'_'+'lst_mnth_trip_count', col+'_'+'lst_mnth_rated_trip_count',col+'_'+'lst_mnth_rated_trip_avg',col+'_'+'lst_mnth_all_trip_rate_avg']
#     summary_old

    col_summary=rate_smry.merge(summary_old, on=[col] , how='left')
#     col_summary
    summary_set.append(col_summary)
```

```python
driver_summary = summary_set[0]
customer_summary = summary_set[1]
```

```python
#Attach Driver Stats with train and test datasets
trip_train_clean_jnd = trip_train_clean.merge(driver_summary, how='left', on ='driver_id')
trip_test_jnd = trip_test.merge(driver_summary, how='left', on ='driver_id')
```

```python
data_combined = [trip_train_clean_jnd,trip_test_jnd]
```

```python
(calendar.weekday(2014, 1, 19) + 1) % 7
```

```python
# Duration in Hours
for dataset in data_combined:
    dataset['creation_date'] = pd.to_datetime(dataset['creation_date'])
    #Generate Trip duration in hours feature
    dataset['duration_time_hours'] = dataset['duration_time']/60
    
    #Generate Trip time feature
    
    dataset['trip_month']=dataset['creation_date'].dt.month
    
    dataset['trip_time']=dataset['creation_date'].apply(timeOfDay,1)
    
    dataset['trip_hour']=dataset['creation_date'].dt.hour
    
    dataset['trip_day']=dataset['creation_date'].dt.day
    
    # Extract Speed (KM/Hour) from time and Distance
    dataset['speed'] = dataset['distance_travelled']/dataset['duration_time_hours']
    
```

#### Standardize numeric values 

```python
#Use MinMax scaler
for dataset in data_combined:
    #Normalize other columns
    colnorm = [ 'estimated_distance', 'distance_travelled',
       'distance_travelled_while_moving', 'estimated_duration',
       'wait_time_initial', 'wait_time_in_journey', 'estimated_price', 'price',
        'duration_time', 'driver_id_trip_count',
       'driver_id_rated_trip_count', 'driver_id_rated_trip_avg',
       'driver_id_all_trip_rate_avg', 'driver_id_lst_mnth_trip_count',
       'driver_id_lst_mnth_rated_trip_count',
       'driver_id_lst_mnth_rated_trip_avg',
       'driver_id_lst_mnth_all_trip_rate_avg', 'duration_time_hours', 'speed']
    for col in colnorm:
        dataset[col]=dataset[col].astype(float)
        dataset[col] = dataset[col].fillna(dataset[col].mean())
        min_max_scaler = preprocessing.MinMaxScaler()
        x_scaled = min_max_scaler.fit(dataset[[col]])
        dataset[col] = x_scaled.transform(dataset[[col]])
```

#### Use One-hot encoding for categorical variables

```python
# one-hot encoding for category features
cat_cols = ['booking_source','car_type', 'trip_month', 'trip_time', 'trip_hour', 'trip_day', 
       'speed_bin', 'trip_weekday']
trip_train_clean_jnd = OneHotEncoderfunc(trip_train_clean_jnd, cat_cols)
trip_train_clean_jnd= trip_train_clean_jnd.loc[:, (trip_train_clean_jnd == 0).sum() < len(trip_train_clean_jnd) *(99.975/100)]
```

#### Prepare dataset for training

```python
trip_train_clean_jnd.columns = [x.lower() for x in trip_train_clean_jnd.columns]
trip_train_clean_jnd.reset_index(drop=True )

train_x = trip_train_clean_jnd.drop(['rating','was_rated','customer_id', 'driver_id', 'creation_date'],axis=1)
train_y = trip_train_clean_jnd['was_rated']
```

#### Use Extra Trees algorithm togenerate feature importance 

```python
##Extra Tree Classifier for feature importance
# train_x.drop(['driver_id_all_trip_rate_avg','is_cancelled'],axis=1, inplace=True)
et = ExtraTreesClassifier(n_estimators = 100, random_state = 100, min_samples_leaf = 3, n_jobs = -1)

et.fit(train_x, train_y)

X= train_x
y= train_y
importances = et.feature_importances_
std = np.std([tree.feature_importances_ for tree in et.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d : %s (%f)" % (f + 1, indices[f],X.columns[indices[f]], importances[indices[f]]))

featuretoplot =60# number of features to plot, use X.shape[1] if you want to plot all features

#Plot the feature importances of the forest
plt.figure(figsize=(20, 10), dpi=80)
plt.title("Feature importances")
#plt.bar(range(X.shape[1]), importances[indices],color="r", yerr=std[indices], align="center")
plt.bar(range(featuretoplot), importances[indices][0:featuretoplot],color="r", yerr=std[indices][0:featuretoplot], align="center")
#plt.xticks(range(X.shape[1]), indices)
plt.xticks(range(featuretoplot), X.columns[indices[0:featuretoplot]])
plt.xticks(fontsize=10, rotation=90,size=7)
#plt.xlim([-1, X.shape[1]])
plt.xlim([-1, featuretoplot])
plt.show()
```

#### Split the training dataset into 70% Training  and 30% Validation
